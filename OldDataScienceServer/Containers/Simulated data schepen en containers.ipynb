```python
import pandas as pd
import numpy as np
import random
```

```python
aantal_boten = 1

#alphenaar:lengte = 6,breedte = 3, hoogte = 2, bokkenboot: lengte = 8,breedte = 5, hoogte = 3
boten = [["alphenaar",6,3,2],["bokkenboot",8,5,3],["pinker",10,4,4]]

rand_stack = []
rand_stack_positie = []
rand_prio = []
boot_id = []
boot_type = []
container_id = []
aantal_containers = 0
prio = [True, False]

for i in range(aantal_boten):
    rand_boot = random.randrange(len(boten))
    stacks = boten[rand_boot][1] * boten[rand_boot][2] 
    containers = stacks * boten[rand_boot][3]
    
    #bootnaam, container_id, stack_id en stack posities genereren 
    boot_id.extend([str(i) for j in range(containers)])
    boot_type.extend([boten[rand_boot][0] for i in range(containers)])
    
    stacks_id = [i for i in range(stacks)]
    stacks_id = stacks_id * boten[rand_boot][3]
    container_id.extend([i + aantal_containers for i in range(containers)])
    stack_posities = [i for i in range(boten[rand_boot][3])] * stacks
    
    #random prio, stack en stack positie genereren
    for i in range(containers):
        #stack
        rand_stack_index = random.randrange(len(stacks_id))
        rand_stack.append(stacks_id[rand_stack_index])
        del stacks_id[rand_stack_index]

        #stack positie
        rand_stack_positie_index = random.randrange(len(stack_posities))
        rand_stack_positie.append(stack_posities[rand_stack_positie_index])
        del stack_posities[rand_stack_positie_index]

        #prio
        rand_prio_index = random.randrange(len(prio))
        rand_prio.append(prio[rand_prio_index])   
    
    aantal_containers = aantal_containers + containers 

#dataframe maken
df = pd.DataFrame(data={'container_id': container_id, 'boot_id': boot_id,'boot_type':boot_type, 'stack_id':rand_stack, 'stack_positie': rand_stack_positie, 'prio': rand_prio})
df
```

```python
dict_df = df.to_dict('list')
```
