```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from pipetorch import DFrame
```

```python
#Import Datasets
df = pd.read_csv("Generated Datasets/ingr_pivot_lunch_diner_noten.csv").set_index("recipe")
tags = pd.read_csv("Datasets Foodboost/tags.csv")

#Get rid of ingredients that aren't used at least 40 times.
ValidColumns = df.loc[:,df.sum() > 40][df.sum(axis=1) != 0]
```

```python
recepten = list(df.columns)
tags = tags[tags['recipe'].str.contains('|'.join(recepten))]
tags_pivot = tags.pivot_table(values="recipe", index= tags.recipe, columns="tag", aggfunc="count")
tags_pivot= tags_pivot.fillna(0) 
tags_ValidColumns = tags_pivot.loc[:,tags_pivot.sum() > 20][tags_pivot.sum(axis=1) != 0]
tags_ValidColumns
```

# keukens voorspellen

```python
keukens = ['amerikaans','aziatisch','chinees','engels','frans','grieks','hollands','indiaas','indonesisch','italiaans', 'japans','marokkaans','mediterraan', 'mexicaans', 'midden-oosters','scandinavisch','spaans','thais','zuid-amerikaans']
df1 = tags_ValidColumns[keukens]
df1["sum"] = df1[keukens].sum(axis = 1)
df1=df1[df1["sum"] == 1]
df1
```

```python
tags2 = tags[tags['tag'].str.contains('|'.join(keukens))]
recepten_1_tag = list(df1.index)
tags2 = tags2[tags2['recipe'].isin(recepten_1_tag)]
tags2 = tags2.iloc[:,1:]
tags2.sort_values('recipe')
```

```python
tags2_ingr = pd.merge(df,tags2, how= 'inner',left_on= 'recipe', right_on = 'recipe')
tags2_ingr = tags2_ingr.set_index('recipe')
#tags2_ingr_copy = tags2_ingr.copy()
#even_smaller_df = tags2_ingr_copy.loc[:,tags2_ingr_copy.sum() > 10][tags2_ingr_copy.sum(axis=1) != 0]
#even_smaller_df
#{k: v for k, v in sorted(d.items(), key=lambda item: item[1])}
i=0
rand_dict = {}
while i < 3:
    gerecht = str(np.random.choice(list(df.index)))
    mening = int(input(str(i) + ': ' + gerecht))
    rand_dict[gerecht] = mening
    i=i+1
rand_dict
```

```python
tags2_ingr
```

```python
X = tags2_ingr.iloc[:,:-1]
y = tags2_ingr.iloc[:,-1]
```

```python
#Generate X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)#, random_state=42)
```

```python
#Import Decision Tree Classifier
from sklearn import tree

#Create and fit Decision Tree Classifier
clf = tree.DecisionTreeClassifier(max_depth = 3).fit(X_train, y_train)

#Create predictions of Decision Tree Classifier
clf_y_pred =  clf.predict(X_test)

#Print confusion matrix and classification report
print(confusion_matrix(y_test, clf_y_pred))
print(classification_report(y_test, clf_y_pred))
```

```python
#Import K Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

#Create and fit K Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors=4).fit(X_train, y_train)

#Create predictions of K Neighbors Classifier
knn_y_pred =  knn.predict(X_test)

#Print confusion matrix and classification report
print(confusion_matrix(y_test, knn_y_pred))
print(classification_report(y_test, knn_y_pred))
```

```python
#Import Logistic Regression
from sklearn.linear_model import LogisticRegression

#Create and fit Logistic Regression
lgr = LogisticRegression(multi_class='multinomial', solver = 'lbfgs').fit(X_train, y_train)

#Create predictions of Logistic Regression
lgr_y_pred =  lgr.predict(X_test)

#Print confusion matrix and classification report
print(confusion_matrix(y_test, lgr_y_pred))
print(classification_report(y_test, lgr_y_pred))
```

```python
tags_ValidColumns_aziatisch = tags_ValidColumns[tags_ValidColumns["aziatisch"] == 1]
aziatischRecipes = tags_ValidColumns_aziatisch.index
aziatischIngredients = ValidColumns[ValidColumns.index.isin(aziatischRecipes.to_list())]
aziatischIngredients['y'] = 0
aziatischIngredients
```

```python
tags_ValidColumns_italiaans = tags_ValidColumns[tags_ValidColumns["italiaans"] == 1]
italiaansRecipes = tags_ValidColumns_italiaans.index
italiaansIngredients = ValidColumns[ValidColumns.index.isin(italiaansRecipes.to_list())]
#italiaansIngredients = italiaansIngredients.iloc[0: len(aziatischIngredients),:]
italiaansIngredients['y'] = 1
italiaansIngredients
```

```python
data = pd.concat([italiaansIngredients,aziatischIngredients])
data = data.loc[:,data.sum() != 0][data.sum(axis=1) != 0]
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
data
```

```python
#Import Logistic Regression
from sklearn.linear_model import LogisticRegression

#Create and fit Logistic Regression
lgr = LogisticRegression(multi_class='multinomial', solver = 'lbfgs').fit(X_train, y_train)

#Create predictions of Logistic Regression
lgr_y_pred =  lgr.predict(X_test)

#Print confusion matrix and classification report
print(confusion_matrix(y_test, lgr_y_pred))
print(classification_report(y_test, lgr_y_pred))
```

```python
#Import K Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

#Create and fit K Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors=4).fit(X_train, y_train)

#Create predictions of K Neighbors Classifier
knn_y_pred =  knn.predict(X_test)

#Print confusion matrix and classification report
print(confusion_matrix(y_test, knn_y_pred))
print(classification_report(y_test, knn_y_pred))
```
