```python
import numpy as np
import random

class TestEnv():
  
    def __init__(self, **kwargs):
        self.width = 6
        self.height = 6   
        self.priorties=[1,2]
   
    def action_to_coords( self, action ):
        return [ action % self.width, int( action / self.width ) ] 

    def is_empty_contraint( self, state, x, y ):
        return state[y][x] == 0
    
    def is_not_floating_constraint( self, state, x, y ):
        if y == 0:
            return True
        if state[y-1][x] != 0:
            return True
        return False
    
    def is_on_top_of_equal_or_higher_priority( self, state, x, y,  container_prio):
        if y == 0:
            return True
        if container_prio >= state[y-1][x]:
            return True
        return False

    def get_possible_actions( self, state,container_prio ):
        actions=[]
        for y in range(self.height):
            for x in range(self.width):
                action = y * self.width + x
                if ( self.is_empty_contraint( state, x, y ) == True and 
                     self.is_not_floating_constraint( state, x, y ) == True and
                     self.is_on_top_of_equal_or_higher_priority( state, x,y,container_prio ) ==True ):
                    actions.append( action )
        return actions 

    def choose_action(self, container_prio, state):
        possible_actions = self.get_possible_actions( state,container_prio )
        
        print("possible actions: ")
        if len(possible_actions) ==0:
            print("None")
            return -1
        for i in possible_actions:
            print(str(i), end=" ")
        print("")
        
        
        action = random.choice( possible_actions )
        
        coords = self.action_to_coords( action )
        print("choosing action: " + str(action) + " coords: "+str(coords[0])+","+str(coords[1]))

        return action

    def step(self, state, container_prio, action):  
        new_state = state
        
        coords = self.action_to_coords( action )
        new_state[ coords[1] ][ coords[0] ] = container_prio
        reward = -1            
        info = {}
        return new_state, reward, info

    def reset(self):
        return np.zeros( ( self.width, self.height ) )
          
    def render(self, state, action, reward):
        print("new state:")
        for y in range(self.height):
            for x in range(self.width):
                print(str(int(state[self.height-1-y][x])),end=" ")
            print("")
        pass
    
env = TestEnv()

# initialize the environment
state = env.reset()

done = False
step = 0
while not done:  
    
  container_prio = random.choice(env.priorties)

  print("------- step " + str( step ) + " --------" )
  print("incoming container with priority " + str(container_prio) )

  # choose a random action
  action = env.choose_action( container_prio, state )

  if action == -1:
        done = True
  else:
      # take the action and get the information from the environment
      new_state, reward, info = env.step(state, container_prio, action)

      state=new_state

      # show the current position and reward
      env.render( state=new_state, action=action, reward=reward)

      step = step + 1
  
  if done == True:
        print("")
        print("Done")
        break
    
  if (step > 100 ):
    break
```
