# Tags dataset inladen en duplicates eruit filteren

```python
# Het inladen van de tags.csv dataframe en hiervan de duplicates eruit filteren. Alleen als een recipe dezelfde tags heeft.
df_tag = pd.read_csv('Dataset Foodboost/tags.csv')
df_tag.drop_duplicates(subset= ['recipe', 'tag'], keep=False, inplace=True)
df_tag
```

# De lekkere recepten met bijbehorende tags

```python
# Alle recepten die lekker zijn laten zien met bijbehorende tags
tags = list()
for recipe in lekker_recepten["recipe"].to_list():
    recipe_tags = df_tag.query("recipe=='" + recipe + "'")
    tags += recipe_tags["tag"].to_list()
    print(str(recipe_tags))
```

# Het aantal keer dat een tag voorkomt optellen

## Jaccard-coëfficient toepassen

### Jaccard-coëfficiënt, is een statistiek die wordt gebruikt voor het meten van de gelijkenis en diversiteit van steekproefreeksen.

```python
# De aantallen van de tags optellen
from collections import Counter

tag_counts = Counter(tags)
print(tag_counts)
```

```python
# Dictionary aanmaken om de scores van elk recept op te slaan
recipe_scores = {}
# Elk recept langsgaan
for recipe in lekker_recepten["recipe"].to_list():
    # Vraag de df_tag dataframe om de tags voor elk recept die langs komt
    recipe_tags = df_tag.query("recipe=='" + recipe + "'")
    # Elk recept die langs komt start met een score van 0
    score = 0
    # Ga elke tag langs voor het huidige recept
    for tag in recipe_tags["tag"].to_list():
        # Voeg de telling van de huidige tag toe aan de score voor het recept
        score += tag_counts[tag]   
    # Voeg de score van het huidige recept toe aan derecipe_scores dictionary
    recipe_scores[recipe] = score
```

# Geef me de top 10 van hoogste score

## Nadat ik 200 recepten op wel of niet lekker heb gerate, kon ik de tags van de lekkere recepten bij elkaar optellen. Hierna heb ik een lijst van top 10 recepten gemaakt van de recepten die ik al als lekker gerate had.

```python
# Sorteer de gescoorde recepten van hoog naar laag
sorted_recipe_scores = sorted(recipe_scores.items(), key=lambda x: x[1], reverse=True)

# Print de top 10 hoogst gescoorde recepten
for recipe, score in sorted_recipe_scores[:10]:
    print(f"{recipe}: {score}")
```
