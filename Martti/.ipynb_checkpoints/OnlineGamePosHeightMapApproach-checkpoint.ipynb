```python
import numpy as np
import pandas as pd
import time
from IPython.display import clear_output
#from numba import njit
```

```python
#Checks if current position is not floating.
def CheckBelow(Environment, X, Y, Z):
    return Z == 0 or (Environment[X,Y,:Z] != '0').all()

#Checks if current position has container.
def CheckPos(Environment, Size, X, Y, Z):
    if not ExceedsLot(Size, X, Y, Z):
        return Environment[X, Y, Z] != '0'
    else:
        return False

#Checks if coordinates exceed lot.
def ExceedsLot(Size, X, Y, Z):
    return Size[2] <= Z or Size[1] <= Y or Size[0] <= X or Z < 0 or Y < 0 or X < 0

#Check whether container is in the middle of other containers.
def CheckMiddle(Environment, Size, X, Y, Z):
    return CheckPos(Environment, Size, X, Y - 1, 0) and CheckPos(Environment, Size, X, Y + 1, 0)
    
```

```python
def GenerateHeightMap(Environment):
    return (Environment != '0').sum(axis=2)

def GeneratePrioMap(Environment):
    HeightMap = GenerateHeightMap(Environment)
    tempMapArr= np.empty(Environment.shape[0:2],dtype='<U1')
    it = np.nditer(HeightMap,["f_index","multi_index"])
    for i in it:
        tempMapArr[it.multi_index[0]][it.multi_index[1]] = Environment[it.multi_index[0], it.multi_index[1], i-1]
    return tempMapArr

def GenerateLegalMap(Environment):
    HeightMap = GenerateHeightMap(Environment)
    legalHeightMove = HeightMap != Environment.shape[2]
    heightNorthandSouth = NorthSouthCheck(HeightMap)
    return np.logical_and(legalHeightMove, heightNorthandSouth)
    
def NorthSouthCheck(HeightMap):
    tempMapArr= np.empty(HeightMap.shape,dtype=bool)
    it = np.nditer(HeightMap,["f_index","multi_index"])
    for i in it:
        tempMapArr[it.multi_index[0]][it.multi_index[1]] = getHeight(HeightMap,it.multi_index[0],it.multi_index[1]+1) == 0 or getHeight(HeightMap,it.multi_index[0],it.multi_index[1]-1) == 0
    return tempMapArr
    
def getHeight(heightm,x,y):
    if x >= 0 and y >= 0 and x < heightm.shape[0] and y < heightm.shape[1]:
        return heightm[x,y]
    else:
        return 0
    
def AddContainer(Environment,X,Y,State):
    TempEnv = Environment.copy()
    HeightMap = GenerateHeightMap(Environment)
    Z = HeightMap[X,Y]
    LegalMap = GenerateLegalMap(Environment)
    #print(LegalMap,X,Y)
    if LegalMap[X,Y]:
        TempEnv[X,Y,Z] = State
    #else:
        #print("FAIL")
        #input()
    return TempEnv



```

```python
#Scores board based on priority :)
def scoreEnvironment(Environment):
    result = 0
    for x in range(Environment.shape[0]):
        for y in range(Environment.shape[1]):
            for z in range(Environment.shape[2]):
                if(Environment[x,y,z] == 'H'):
                    #print("yes")
                    #result += (Environment[x,y,z+1:] == 'L').sum()
                    north = (Environment[x,y:,:] == 'L').sum()
                    south = (Environment[x,:y+1,:] == 'L').sum()
                    
                    result += -np.min(north + south)
    return result


#Returns cleared environment
def clearEnvironment(Environment):
    return np.array([[['0'] * Environment.shape[2]] * Environment.shape[1]] * Environment.shape[0])

def singleLegalSpotRemoved(Env1,Env2):
    firstEnvSpots = np.sum(GenerateLegalMap(Env1))
    secondEnvSpots = np.sum(GenerateLegalMap(Env2))
    #print("FirstEnv: %i, SecondEnv: %i"%(firstEnvSpots,secondEnvSpots))
    if np.array_equiv(Env1,Env2):
        return True
    if (firstEnvSpots - secondEnvSpots) <= 1:
        return False
    return True

#Generate status of board in Boolean array
def GenerateStatusDataset(Environment,dataType):
    return np.concatenate([np.array(GenerateLegalMap(Environment),dtype=np.float32).flatten(),np.array(GenerateHeightMap(Environment), dtype=np.float32).flatten()])
#    return np.concatenate([np.array([dataType == 'H'], dtype=np.float32),np.array(GenerateHeightMap(Environment), dtype=np.float32).flatten(), np.array(GenerateLegalMap(Environment),dtype=np.float32).flatten()])


def renderBalk(value,maxvalue,barlength,reverse=False):
    bars = int((value)/maxvalue*barlength)
    if reverse:
        return str('#'*bars) + str('_'*(barlength-bars))
    else:
        return str('#'*(barlength-bars)) + str('_'*bars)
```

```python
import torch as T
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
class DeepQNetwork(nn.Module):
    def __init__(self, lr, input_dims, fc1_dims, fc2_dims, n_actions):
        super(DeepQNetwork, self).__init__()
        self.input_dims = input_dims
        self.fc1_dims = fc1_dims
        self.fc2_dims = fc2_dims
        self.n_actions = n_actions
        self.fc1 = nn.Linear(*self.input_dims, self.fc1_dims)
        self.fc2 = nn.Linear(self.fc1_dims, self.fc2_dims)
        self.fc3 = nn.Linear(self.fc2_dims, self.n_actions)
        self.optimizer = optim.Adam(self.parameters(),lr=lr)
        self.loss = nn.MSELoss()
        self.device = T.device('cuda:0' if T.cuda.is_available() else 'cpu')
        self.to(self.device)
    
    def forward(self, state):
        x = F.relu(self.fc1(state))
        x = F.relu(self.fc2(x))
        actions = self.fc3(x)
        
        return actions
```

```python
class Agent():
    def __init__(self, gamma, epsilon, lr, input_dims, batch_size, n_actions, fc1_dims=128,fc2_dims=128, max_mem_size=100000, eps_end=0.01, eps_dec=5e-4):
        self.gamma = gamma
        self.epsilon = epsilon
        self.eps_min = eps_end
        self.eps_dec = eps_dec
        self.lr = lr
        self.action_space = [i for i in range(n_actions)]
        self.mem_size = max_mem_size
        self.batch_size = batch_size
        self.mem_cntr = 0
        
        self.Q_eval = DeepQNetwork(self.lr, n_actions=n_actions, input_dims=input_dims, fc1_dims = fc1_dims, fc2_dims=fc2_dims)
        
        self.state_memory = np.zeros((self.mem_size, *input_dims), dtype=np.float32)
        self.new_state_memory = np.zeros((self.mem_size,*input_dims), dtype=np.float32)
        
        self.action_memory = np.zeros(self.mem_size, dtype=np.int32)
        self.reward_memory = np.zeros(self.mem_size, dtype=np.float32)
        self.terminal_memory = np.zeros(self.mem_size, dtype=np.bool)
        
    def store_transition(self, state, action, reward, state_, done):
        index = self.mem_cntr % self.mem_size
        self.state_memory[index] = state
        self.new_state_memory[index] = state_
        self.reward_memory[index] = reward
        self.action_memory[index] = action
        self.terminal_memory[index] = done
        
        self.mem_cntr += 1
    
    def choose_action(self, observation):
        if np.random.random() > self.epsilon:
            state = T.tensor([observation]).to(self.Q_eval.device)
            actions = self.Q_eval.forward(state)
            action = T.argmax(actions).item()
        else:
            action = np.random.choice(self.action_space)
        return action
    
    def learn(self):
        if self.mem_cntr < self.batch_size:
            return
        
        self.Q_eval.optimizer.zero_grad()
        
        max_mem = min(self.mem_cntr, self.mem_size)
        batch = np.random.choice(max_mem, self.batch_size, replace=False)
        
        batch_index = np.arange(self.batch_size, dtype=np.int32)
        
        state_batch = T.tensor(self.state_memory[batch]).to(self.Q_eval.device)
        new_state_batch = T.tensor(self.new_state_memory[batch]).to(self.Q_eval.device)
        reward_batch = T.tensor(self.reward_memory[batch]).to(self.Q_eval.device)
        terminal_batch = T.tensor(self.terminal_memory[batch]).to(self.Q_eval.device)
        
        action_batch = self.action_memory[batch]
        
        q_eval = self.Q_eval.forward(state_batch)[batch_index, action_batch]
        q_next = self.Q_eval.forward(new_state_batch)
        q_next[terminal_batch] = 0.0
        
        q_target = reward_batch + self.gamma * T.max(q_next, dim=1)[0]
        
        loss = self.Q_eval.loss(q_target, q_eval).to(self.Q_eval.device)
        loss.backward()
        self.Q_eval.optimizer.step()
        
        self.epsilon = self.epsilon - self.eps_dec if self.epsilon > self.eps_min else self.eps_min
    
```

```python
#TODO: change Epsilon End based on size (maybe done idk gotta check if it works well first tho)
def trainNetwork(size, bs, learnrate, games, rewardamt = 3, punishamt = -10, SmoothingFactor = 200, fc1_dims=128,fc2_dims=128):
    Size = size
    #Generate Environment
    c =  [[['0'] * Size[2]] * Size[1]] * Size[0]
    maxContainers = Size[0]*Size[1]*Size[2]
    ObservationSize = Size[0]*Size[1]*2 #Set observation space size
    #actionSize = Size[0]*Size[1] #Set action space size
    actionSize = Size[0]+Size[1]+Size[2]
    Speelveld = np.array(c)




    #Create agent with properties
    agent = Agent(gamma=0.99, epsilon=1.0, batch_size=bs, fc1_dims=fc1_dims,fc2_dims=fc2_dims, n_actions=actionSize, eps_end=1/maxContainers, input_dims = [ObservationSize], lr=0.001)

    scores, eps_history, avg_scores = [], [], []

    #Number of games
    n_games = games

    for i in range(n_games):
        score = 0
        done = False
        Speelveld = clearEnvironment(Speelveld)
        step = 0
        containerList = []

        timings = []
        #containerList = [np.random.choice(['L','H'])]*actionSize
        for o in range(maxContainers):
            containerList.append(np.random.choice(['L','H']))
        #print(containerList)
        #containerList = ['H','L','L','H']

        MemoryDict = {"newContainerType":[],
                      "Speelveld":[],
                      "newSpeelveld":[],
                      "action":[],
                      "reward":[],
                      "done":[]}


        while not done:
            reward = 0
            timings = []
            step+=1
            newContainerType = containerList[-1]
            containerList.pop()
            #Timing
            timings.append(time.time_ns())
            Observation = GenerateStatusDataset(Speelveld,newContainerType)
            timings.append(time.time_ns())
            #print(Observation)

            #Neural Network
            timings.append(time.time_ns())
            action = agent.choose_action(Observation)
            timings.append(time.time_ns())
            #Insert while loop waar alle opties geprobeerd worden.
            nnpred = np.array([0]*Size[1] * Size[0])
            nnpred[action] = 1

            posX, posY = np.where(nnpred.reshape((Size[0],Size[1]))==1)

            timings.append(time.time_ns())
            newSpeelveld = AddContainer(Speelveld, int(posX), int(posY), newContainerType)
            timings.append(time.time_ns())

            #Rewarding or punishing agent:
            timings.append(time.time_ns())
            if singleLegalSpotRemoved(Speelveld,newSpeelveld):
                containerList.append(newContainerType)
                reward = punishamt
                done = True
            else:
                reward = rewardamt
            timings.append(time.time_ns())

            #Exit environment if there's been 2 too many step s (No need to punish, since it already punishes itself above)
            if(step >= Size[0]*Size[1]*Size[2]*2 or not containerList):
                done = True

            score += reward

            #Append to memory lists
            MemoryDict["newContainerType"].append(newContainerType)
            MemoryDict["Speelveld"].append(Speelveld)
            MemoryDict["newSpeelveld"].append(newSpeelveld)
            MemoryDict["action"].append(action)
            MemoryDict["reward"].append(reward)
            MemoryDict["done"].append(done)

            Speelveld = newSpeelveld


            eps_history.append(agent.epsilon)
            avg_score = np.mean(scores[-100:])
        scores.append(score)
        avgReward = np.mean(MemoryDict["reward"])
        #avgReward = 0
        #avgReward = scoreEnvironment(Speelveld)
        #print(MemoryDict.values())
        #print("Episode %i score %.2f average score %.2f epsilon %.2f"%(i,score,avg_score,agent.epsilon))


        if not i % 10:
            #print(int(len(timings)/2))

            clear_output(wait=True)
            #for a in range(int(len(timings)/2)):
            #    print("Timing %i lasted for: %i"%(a,timings[a*2+1] - timings[a*2]))
            
            #Calculate the maximum value based on the smoothing factor.
            maxRenderedEpisode  = int(len(scores)/SmoothingFactor)*SmoothingFactor

            #Create list of smoothed scores
            smthscr = np.median(np.reshape(scores[:maxRenderedEpisode],(-1,SmoothingFactor)),axis=1)
            
            
            #Breaks simulation if neural network is fully trained
            if np.mean(smthscr[-5:]) == maxContainers*reward and np.std(smthscr[-5:]) < 0.3:
                break
            
            unique, counts = np.unique(Speelveld.flatten(), return_counts=True)
            occurrences = dict(zip(unique, counts))
            if '0' in occurrences:
                print("Empty spots: %i\t"%(occurrences['0'])+renderBalk(occurrences['0'],Size[0]*Size[1]*Size[2],40,reverse=True))
            else:
                print("No empty spots! Almost done!")

            print("Train progress:\t" + renderBalk(i,n_games,40,reverse=True))
            print("Episode %i/%i score %.2f avgReward %.2f epsilon %.2f"%(i,n_games,np.mean(scores[-100:]),avgReward,agent.epsilon),flush=True)



        for l in range(len(MemoryDict["newContainerType"])):
            agent.store_transition(GenerateStatusDataset(MemoryDict["Speelveld"][l],MemoryDict["newContainerType"][l]), MemoryDict["action"][l], avgReward + MemoryDict["reward"][l], GenerateStatusDataset(MemoryDict["newSpeelveld"][l],MemoryDict["newContainerType"][l]), MemoryDict["done"][l])
        agent.learn()

        #print(Speelveld)

        avg_scores.append(avgReward)
    #    x = [i+1 for i in range(n_games)]

    #    plot_learning_curve(x, scores, eps_history, filename)


    
    #SmoothingFactor = 200
    
    #Calculate the maximum value based on the smoothing factor.
    maxRenderedEpisode  = int(len(scores)/SmoothingFactor)*SmoothingFactor
    
    #Create list of x Axis Labels so they match the smoothing factor
    xAxisList = np.arange(0,maxRenderedEpisode,SmoothingFactor)
    
    #Create list of smoothed scores
    SmoothedScores = np.median(np.reshape(scores[:maxRenderedEpisode],(-1,SmoothingFactor)),axis=1)
    
    #Calculate the maximum episode that 
    MaxEpisode = int(np.argmax(SmoothedScores) * 1.5)
    
    
    #plt.plot(np.array(range(int(MaxEpisode/SmoothingFactor)))*SmoothingFactor,SmoothedScores,color="navy",label="Median score over "+ str(SmoothingFactor) + " games")
    plt.plot(xAxisList[:MaxEpisode] ,SmoothedScores[:MaxEpisode],color="navy",label="Median score over "+ str(SmoothingFactor) + " games")
    plt.axhline(y = Size[0]*Size[1]*Size[2]*3, color = 'r', linestyle = '--',label="Maximum score")

    plt.legend()
    plt.title("Training score reinforcement learning \n(%ix%ix%i HeightMap Batch Size %i, learning rate %f)"%(Size[0],Size[1],Size[2],agent.batch_size,agent.lr))
    plt.xlabel("Episodes")
    plt.ylabel("Score")

    plt.savefig('Graphs/Martti-%ix%ix%i-b%i-lr%f.png'%(Size[0],Size[1],Size[2],agent.batch_size,agent.lr),dpi=300)
    plt.cla()
    T.save(agent.Q_eval.state_dict(),"Graphs/MarttiWeights-%ix%ix%i-b%i-lr%f.pt"%(Size[0],Size[1],Size[2],agent.batch_size,agent.lr))
    
```
